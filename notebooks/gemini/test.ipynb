````python
import os
import json
import pandas as pd
from tqdm import tqdm

# Process single video
def process_video(video_path, output_csv):
    try:
        uploaded_video = upload_video(video_path)
        
        response = client.models.generate_content(
            model=model_name,
            contents=[uploaded_video, sexism_prompt]
        )

        print(response)
        
        result = json.loads(response.text.strip().strip("```json").strip("```"))
        print(result)
        description = result.get("description", "ERROR")
        label = result.get("label", "ERROR")
        analysis = result.get("analysis", "ERROR")
        
        # Create single row dataframe
        df_result = pd.DataFrame({
            'description': [description],
            'label': [label],
            'analysis': [analysis]
        })
        
        # Save to CSV
        df_result.to_csv(output_csv, index=False)
        print(f"💾 Saved results to: {output_csv}")
        
    except Exception as e:
        print(f"❌ Error processing video: {e}")
        df_result = pd.DataFrame({
            'description': ['ERROR'],
            'label': ['ERROR'],
            'analysis': ['ERROR']
        })
        df_result.to_csv(output_csv, index=False)

# Example usage
video_path = "/Users/moiz.ali/Downloads/EXIST 2025 Dataset V0.3/EXIST 2025 Videos Dataset/test/videos/7253209738391866670.mp4"  # Replace with your video path
output_csv = "420072_fp.csv"
process_video(video_path, output_csv)
````
