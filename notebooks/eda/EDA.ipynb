```python
# only select annotators and gender_annotator columns
annotators_df = df[['annotators', 'gender_annotators','labels_task3_1']]

annotators_df.head()
```

```python
# Flatten the list and count occurrences of each label
flat_labels = [label for sublist in df['labels_task3_1'] for label in sublist]
label_counts = Counter(flat_labels)

# Print the count of each label
print(label_counts)
```

```python
# Create a dictionary to store gender mapping
annotator_gender_map = {}

# Iterate through the dataset
for annotators, genders in zip(df['annotators'], df['gender_annotators']):
    for annotator, gender in zip(annotators, genders):
        if annotator in annotator_gender_map and annotator_gender_map[annotator] != gender:
            print(f"⚠️ Inconsistency found for {annotator}: {annotator_gender_map[annotator]} vs {gender}")
        annotator_gender_map[annotator] = gender

# Print the final mapping
print("Annotator Gender Mapping:")
print(annotator_gender_map)

df_annotator = pd.DataFrame(list(annotator_gender_map.items()), columns=['Annotator', 'Gender'])

df_annotator = df_annotator.sort_values(by="Gender")
df_annotator
```

```python
# Create a dictionary to store "YES" and "NO" counts per annotator

# ONLY LOOK AT THE ONES WITH CONFLICT OPINIONS
annotator_counts = {}

# Iterate through the dataset
for annotators, labels in zip(df['annotators'], df['labels_task3_1']):
    for annotator, label in zip(annotators, labels):
        if annotator not in annotator_counts:
            annotator_counts[annotator] = {"YES": 0, "NO": 0,"UNKNOWN":0}
        annotator_counts[annotator][label] += 1

# Convert dictionary to DataFrame
count_df = pd.DataFrame.from_dict(annotator_counts, orient='index').fillna(0)

# Order by gender
custom_order = df_annotator['Annotator'].to_list()
count_df = count_df.loc[custom_order]

# Plot Stacked Bar Chart
count_df.plot(kind="bar", stacked=True, figsize=(10, 6), color=["green", "red","blue"])
plt.xlabel("Annotator")
plt.ylabel("Count of YES and NO")
plt.title("Count of YES and NO Responses per Annotator")
plt.xticks(rotation=45)
plt.legend(title="Response")
plt.show()
```

```python
from collections import Counter
import re
import nltk
from nltk.corpus import stopwords

# make 
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

df_en = df[df['lang'] == 'en']

# Tokenize and count words
all_words = ' '.join(df_en['text']).lower()
words = re.findall(r'\b\w+\b', all_words)  # Extract words
filtered_words = [word for word in words if word not in stop_words]  # Remove stopwords
word_counts = Counter(filtered_words)

# Convert to DataFrame for visualization
word_df = pd.DataFrame(word_counts.most_common(20), columns=['Word', 'Count'])

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=word_df, x='Word', y='Count')
plt.xticks(rotation=45)
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.title("Top 10 Most Frequent Words (Without Stopwords)")
plt.show()
```
